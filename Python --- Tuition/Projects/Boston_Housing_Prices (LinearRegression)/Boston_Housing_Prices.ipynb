# Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Importing the Boston Housing Prices Datasets

```python
from sklearn.datasets import load_boston
boston = load_boston()
boston
```

```python
data = pd.DataFrame(boston.data)
```

```python
data.head()
```

```python
data.columns = boston.feature_names
data.head()
```

```python
data['Prices'] = boston.target
```

```python
data.head(20)
```

## Correlation and plotting it

```python
corr = data.corr()
corr
```

```python
plt.figure(figsize = (20, 10))
sns.heatmap(corr, cmap = "Blues", annot = True)
```

```python
y = data['Prices']
y
```

```python
X = data.drop(['Prices'], axis = 1)
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

## Running the LinearRegression Model!

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
print(y_pred)
```

```python
sns.distplot((y_test - y_pred), bins = 50)
```

```python
print(y_test - y_pred)
```

## Metrics related for evaluating model performance

### Errors
* Mean Absolute Error
* Root Mean Squared Error

```python
from sklearn import metrics
```

```python
absolute = metrics.mean_absolute_error(y_test, y_pred)
```

```python
squared = np.sqrt(absolute)
```

```python
print("Root Mean Squared Error: ")
print(squared)
print("\nMean Absolute Error: ")
print(absolute)
```
