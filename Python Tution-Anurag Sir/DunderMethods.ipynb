# NEGATIVE, POSITIVE AND INVERT DUNDER METHODS

```python
class unary_str:
    def __init__(self, string):
        self.string1 = string
    def __str__(self):
        return self.string1
    def __repr__(self):
        return str(self)
    
    def __neg__(self):
        last_index = len(self.string1) - 1
        i = 0
        list_str = []
        for x in range(len(self.string1)):
            list_str.append(self.string1[x])
        while i <= last_index:
            str_el = list_str[i]
            list_str[i] = list_str[last_index]
            list_str[last_index] = str_el
            i += 1
            last_index -= 1
        
        return list_str
    def __pos__(self):
        return self.__neg__()
    def __invert__(self):
        return self.__neg__()
```

```python
class student:
    def __init__(self, subjects, out_of, marks = None):
        self.subjects_l = subjects
        self.total_marks = out_of
        if len(marks) < len(subjects):
            diff = len(subjects) - len(marks)
            for _ in range(diff):
                marks.append(0)
        elif len(marks) > len(subjects):
            diff = len(marks) - len(subjects)
            marks = marks[:len(subjects)]
        self.marks_l = marks

    def __str__(self):
        return f"{self.subjects_l}" + str(self.total_marks) + f"{self.marks_l}"
    def __repr__(self):
        return str(self)

    def __mul__(self, other):
        if isinstance(other, (int, float)):
            total_marks = self.total_marks.copy()
            marks_l = self.marks_l.copy()
            indexes = len(self.marks_l) - 1
            for _ in range(indexes):
                total_marks[_] *= other
                marks_l[_] *= other
            return student(self.subjects_l.copy(), total_marks, marks_l)
        raise Exception(f"* or *= is not supported between Type 'Student' and {other.__class__.__name__}")
    def __imul__(self, other):
        if isinstance(other, (int, float)):
            indexes = len(self.marks_l) - 1
            for _ in range(indexes):
                self.total_marks[_] = self.total_marks[_] * other
                self.marks_l[_] = self.marks_l[_] * other
            return self
        raise Exception(f"* or *= is not supported between Type 'Student' and {other.__class__.__name__}")

    def __truediv__(self, other):
        if isinstance(other, (float, int)):
            indexes = len(self.marks_l) - 1
            for _ in range(indexes):
                self.total_marks[_] = self.total_marks[_] / other
                self.marks_l[_] = self.marks_l[_] / other
            return str(self.total_marks) + f"{self.marks_l}"
        raise Exception(f"/ or /= is not supported between Type 'Student' and {other.__class__.__name__}")
    def __idiv__(self, other):
        return self.__truediv__(other)
```
